```python
import csv
import json
from bs4 import BeautifulSoup
import requests
import lxml

url = 'https://www.restoclub.ru/spb/ratings'
headers = {'Accept': '*/*',
           'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 YaBrowser/22.11.5.715 Yowser/2.5 Safari/537.36'}
#
# Достаем ссылку на сайт

req = requests.get(url,headers=headers)
src = req.text
```

```python
soup = BeautifulSoup(src,"html.parser")
```

```python
all_next_10_list_href = []
all_next_10_html = soup.find(class_="page-ratings__list").find_all("a")
all_next_10_html
for next_10 in all_next_10_html:
    next_10_href = next_10.get('href')
    if next_10_href[5] == 'r':
        next_10_href = 'https://www.restoclub.ru'+next_10.get('href')
        all_next_10_list_href.append(next_10_href)
all_next_10_list_href
```

```python
all_next_10_name = soup.find_all('h3')
all_next_10_name = all_next_10_name[:-5]
all_next_10_name_text = []
for name in range(len(all_next_10_name)):
    all_next_10_name[name] = all_next_10_name[name].text
all_next_10_name
```

```python
zip_all_catig = dict(zip(all_next_10_name,all_next_10_list_href))
inv_zip_all_catig = dict(zip(all_next_10_list_href,all_next_10_name))
inv_zip_all_catig
```

```python
list_rest = []

for url in all_next_10_list_href:
    # def rest_info_append(url):
    # url = 'https://www.restoclub.ru/spb/ratings/rejting-zagorodnyh-restoranov'
    req = requests.get(url,headers=headers)
    src = req.text
    soup = BeautifulSoup(src,"html.parser")
    name_rest = soup.find(class_="rating-places _full").find_all(class_="search-place-title__name")
    rest_metro = soup.find(class_="rating-places _full").find_all("li", class_ = 'search-place-card__info-item')
    rest_raiting_vh = soup.find(class_="rating-places _full").find_all('div', class_ = 'rating__value very-high')
    rest_raiting_h = soup.find(class_="rating-places _full").find_all('div', class_ = 'rating__value high')

    for name in range(len(name_rest)):
        name_rest[name] = name_rest[name].text

    metros=[]
    for metro in range(len(rest_metro)):
        rest_metro[metro] = rest_metro[metro].text
        if rest_metro[metro] != rest_metro[metro].lower():
            metros.append(rest_metro[metro])
    if url == 'https://www.restoclub.ru/spb/ratings/rejting-zagorodnyh-restoranov':
        metros.insert(1, None)
    if url == 'https://www.restoclub.ru/spb/ratings/rejting-barov-peterburga':
        metros.insert(5, None)
    if url == 'https://www.restoclub.ru/spb/ratings/familytime':
        metros.insert(5, None)    

    rest_raiting = []
    for raiting in range(len(rest_raiting_vh)):
        rest_raiting.append(rest_raiting_vh[raiting].text)
    for rateing in range(len(rest_raiting_h)):
        rest_raiting.append(rest_raiting_h[rateing].text)
    rest_raiting

    cats=[]
    for item in inv_zip_all_catig:
        if url == item:
            for n in range(10):
                cats.append(inv_zip_all_catig[item])


    for rest in range(10):
        dataframe = {}
        dataframe['Name'] = name_rest[rest]
        dataframe['Raiting'] = rest_raiting[rest]
        dataframe['Metro'] = metros[rest]
        dataframe['Category'] = cats[rest]
        list_rest.append(dataframe)
    print(f'{inv_zip_all_catig[url]} внесен')
```

```python
list_rest
```

```python
import pandas
df = pandas.DataFrame(list_rest)
df.to_csv("Spb_res.csv",index=False)
```
